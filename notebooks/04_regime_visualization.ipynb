# Regime Visualization

Appendix notebook for reviewing bull, neutral, and bear market-index regimes.

```python
from pathlib import Path

import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
features_path = ROOT / 'data' / 'features' / 'features_day5_v1.parquet'
features = pd.read_parquet(features_path) if features_path.exists() else pd.DataFrame()
features.shape
```

```python
regime_cols = [
    'cs2_index_regime_bull',
    'cs2_index_regime_neutral',
    'cs2_index_regime_bear',
]
if not features.empty and {'timestamp', *regime_cols}.issubset(features.columns):
    regime = features.copy()
    regime['date'] = pd.to_datetime(regime['timestamp']).dt.date
    timeline = regime.groupby('date')[regime_cols].mean()
    display(timeline.tail(30))
else:
    display(pd.DataFrame(columns=regime_cols))
```

```python
if not features.empty and 'cs2_index_return_30d' in features.columns:
    display(features[['timestamp', 'market_hash_name', 'cs2_index_return_30d']].tail(50))
else:
    display(pd.DataFrame())
```
