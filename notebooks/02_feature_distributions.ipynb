# Feature Distributions

Appendix notebook for checking feature coverage, numeric distributions, and simple correlations.

```python
from pathlib import Path

import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
features_path = ROOT / 'data' / 'features' / 'features_day5_v1.parquet'
features = pd.read_parquet(features_path) if features_path.exists() else pd.DataFrame()
features.shape
```

```python
numeric = features.select_dtypes(include='number') if not features.empty else pd.DataFrame()
coverage = numeric.notna().mean().sort_values().rename('non_null_coverage')
coverage.tail(20)
```
